# District Map Plotter
This is simply meant to plot a district plan. Just check a districting plan with `Repairs.ipynb` and then you can plot it! You will likely have to change the key for the district depending on the shapefile you use.

```python
import geopandas as gpd
import matplotlib.pyplot as plt

plan_gdf = gpd.read_file("NewMexico/currenthouseplan/HD_221619_shapefile/HD_221619.shp")
```

```python
plan_gdf.columns
```

```python
# Dissolve: Group by district, combine other columns by summing up votes
plan_gdf_districts = plan_gdf.dissolve(by = "DISTRICT", aggfunc = 'sum' )
```

```python
# Make empty figure and axes
fig, ax = plt.subplots()

# Plot colors and lines using the same axis
dist_plot = plan_gdf.plot(ax = ax, column = "DISTRICT", cmap ='prism')
plan_gdf_districts.boundary.plot(ax =ax, color = "black", linewidth = 1)

# show your figure
plt.show()
```
